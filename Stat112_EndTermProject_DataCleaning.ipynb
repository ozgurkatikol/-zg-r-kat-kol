# **Data Cleaning**


```python
import numpy as np
import pandas as pd
from statistics import mode

df= pd.read_excel('Ford_Car_Price.xlsx')

# We imported the Python libraries Pandas, NumPy and Statistics to clean and manipulate the dirty data
# Then we read the provided excel file as df
```

```python
# We examined the variables and their data types by using .info() and printing df
df.info()
df
```

```python
# After inspecting we read the excel as df_corrected and set the third row as the header
df_corrected = pd.read_excel('Ford_Car_Price.xlsx', header=2)


# We combined the first few rows and the last few rows of the corrected DataFrame using pd.concat to display a concise summary of the corrected DataFrame
df_corrected_head_tail = pd.concat([df_corrected.head(), df_corrected.tail()])

# We then displayed the combined corrected DataFrame containing the first few and last few rows along with the data types of each column in the corrected DataFrame
df_corrected_head_tail, df_corrected.dtypes

```

```python
# We defined a dictionary that maps the old column names to the new column names
column_names = {
    '##Year': 'Year',
    'Month': 'Month',
    '//Model': 'Model',
    'COLOR': 'Color',
    'Condition': 'Condition',
    'TRANSMISSION': 'Transmission',
    '"FUEL TYPE"': 'Fuel Type',
    '##Mileage': 'Mileage',
    'Engine_Size': 'Engine Size',
    '##Price': 'Price',
    '//**Owner_Type\\': 'Owner Type',
    }



# We renamed the columns of the DataFrame based on the defined mapping
df_corrected.rename(columns=column_names, inplace=True)
df_corrected.columns.values[10] = 'Owner Type'
# We printed the column names to make sure it was working.
df_corrected.columns

# We converted numeric columns to float data type and handled errors by coercing invalid values to NaN
df_corrected['Price'] = pd.to_numeric(df_corrected['Price'], errors='coerce').astype(float)
df_corrected['Mileage'] = pd.to_numeric(df_corrected['Mileage'], errors='coerce').astype(float)
df_corrected['Engine Size'] = pd.to_numeric(df_corrected['Engine Size'], errors='coerce').astype(float)

# We didn't make the "year" column date or integer because the year represents a descriptive data, indicating the model year of the car, in other words.
df_corrected['Year'] = df_corrected['Year'].astype(object)


# We displayed the combined corrected DataFrame containing the first few and last few rows along with the data types of each column in the corrected DataFrame
df_corrected_head_tail = pd.concat([df_corrected.head(), df_corrected.tail()])
df_corrected_head_tail, df_corrected.dtypes
```

```python
# We then inspected the DataFrame to see whether the Non-Null counts changed using .info()
df_corrected.info()
```

```python
# We checked the existance of outliers using .describe()
df_corrected.describe()
```

```python
# We used .duplicated() to find duplicates in the DataFrame and then used .sum() to find the total number of dublicates in the DataFrame
df_corrected.duplicated().sum()
```

```python
# We removed the duplicate rows from the original DataFrame and created a new DataFrame df_no_duplicates
# We used the .copy() to ensure that the new DataFrame is a separate copy
df_no_duplicates = df_corrected.drop_duplicates().copy()


# We removed leading and trailing whitespaces from the object columns
df_no_duplicates['Model'] = df_no_duplicates['Model'].str.strip()
df_no_duplicates['Color'] = df_no_duplicates['Color'].str.strip()
df_no_duplicates['Condition'] = df_no_duplicates['Condition'].str.strip()
df_no_duplicates['Transmission'] = df_no_duplicates['Transmission'].str.strip()
df_no_duplicates['Month'] = df_no_duplicates['Month'].str.strip()
df_no_duplicates['Year'] = df_no_duplicates['Year'].astype(str).str.strip()
df_no_duplicates['Owner Type'] = df_no_duplicates['Owner Type'].str.strip()
df_no_duplicates['Fuel Type'] = df_no_duplicates['Fuel Type'].str.strip()


# We created an string_columns list and then converted string columns to lowercase to ensure that all strings are in the same format
string_columns = ['Year','Month', 'Model', 'Color', 'Condition', 'Transmission', 'Fuel Type', 'Owner Type']
df_no_duplicates.loc[:, string_columns] = df_no_duplicates[string_columns].apply(lambda x: x.str.lower())


# We iterated over each column specified in the string_columns list and printed the value counts for each unique value in that column within the DataFrame to identify typos
for column in string_columns:

    print(f"Value Counts for {column}:\n{df_no_duplicates[column].value_counts()}\n")


```

```python
# Standardizing the format of the string columns

# We corrected specific substrings in string columns
for column in string_columns:

    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace('auto ', 'automatic')
    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace(' u ', 'used')
    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace(' n ', 'new')
    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace('"petrol "', 'petrol')
    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace('"Hybrid"', 'hybrid')
    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace('"Diesel"', 'diesel')
    df_no_duplicates.loc[:, column] = df_no_duplicates[column].str.replace('"Petrol"', 'petrol')


# We corrected u as 'Used' and n as 'New' and capitalized the 'used' and 'new' words
df_no_duplicates['Condition'] = df_no_duplicates['Condition'].replace({'u': 'Used', 'n': 'New'}).str.capitalize()


# We corrected the model names, replaced underscores with hyphens and capitalized each word
df_no_duplicates['Model'] = df_no_duplicates['Model'].str.replace('_', '-').str.title()
df_no_duplicates['Model'] = df_no_duplicates['Model'].str.replace('Focuss', 'Focus').str.title()


# We corrected color names by capitalizing each word
df_no_duplicates['Color'] = df_no_duplicates['Color'].str.capitalize()


# We corrected month names by capitalizing each word
df_no_duplicates['Month'] = df_no_duplicates['Month'].str.capitalize()


# We corrected transmission by replacing 'auto' with 'Automatic' and then capitalized each word
df_no_duplicates['Transmission'] = df_no_duplicates['Transmission'].replace({'auto': 'Automatic'}).str.capitalize()


# We removed quotes from fuel type and then capitalized each word
df_no_duplicates['Fuel Type'] = df_no_duplicates['Fuel Type'].str.replace('"', '').str.capitalize()


# We changed the mistakes in owner type and then capitalized each word
df_no_duplicates['Owner Type'] = df_no_duplicates['Owner Type'].replace({'1st': 'First', '2nd': 'Second', '3rd': 'Third'}).str.capitalize()

# We checked again to see if there were any typo's left
for column in string_columns:

  print(f"Value Counts for {column}:\n{df_no_duplicates[column].value_counts()}\n")
# Checking the unique values in the column of the original dataset
for column in string_columns:
  df_no_duplicates[column].unique()


# We also corrected the indexes of each row(it was in disarray after the dublicates were deleted)
new_index=[list(range(1, 71))]
df_no_duplicates.set_index(new_index, inplace=True)

# We checked the data after standardization
df_no_duplicates_head_tail = pd.concat([df_no_duplicates.head(), df_no_duplicates.tail()])
df_no_duplicates_head_tail
```

```python
# We used .isna() to find how many Na values are in the DataFrame and in which columns
na_count = df_no_duplicates.isna().sum()
na_count
```

```python
# We calculated the mean value of the "Milage" column using np.mean and replaced Na values in that column with the mean value
mean_of_Mileage=np.mean(df_no_duplicates["Mileage"])
df_no_duplicates['Mileage'] = df_no_duplicates['Mileage'].replace(np.NaN,mean_of_Mileage,regex=True)


# We calculated the mean value of the "Engine Size" column using np.mean and replaced Na values in that column with the mean value
mean_of_Engine_Size=np.mean(df_no_duplicates["Engine Size"])
df_no_duplicates['Engine Size'] = df_no_duplicates['Engine Size'].replace(np.NaN,mean_of_Engine_Size,regex=True)


# We calculated the mean value of the "Price" column using np.mean and replaced Na values in that column with the mean value
mean_of_Price=np.mean(df_no_duplicates["Price"])
df_no_duplicates['Price'] = df_no_duplicates['Price'].replace(np.NaN,mean_of_Price,regex=True)


# We found the mode of the "Owner Type" column using mode function from the statistics library and replaced Na values in that column with the mode
mode_of_Owner_Type=mode(df_no_duplicates["Owner Type"])
df_no_duplicates['Owner Type'] = df_no_duplicates['Owner Type'].replace(np.NaN,mode_of_Owner_Type,regex=True)

# We found the mode of the "Color" column using mode function from the statistics library and replaced Na values in that column with the mode
mode_of_Color=mode(df_no_duplicates["Color"])
df_no_duplicates['Color'] = df_no_duplicates['Color'].replace(np.NaN,mode_of_Color,regex=True)


# We checked the na counts again to see if there were any left
na_count = df_no_duplicates.isna().sum()
na_count
```

```python
# We controlled the DataFrame to make sure there were no mistake
df_no_duplicates.describe()
df_no_duplicates.info()
```

```python
# We exported the cleaned DataFrame to a new excel file named "CleanedData" for further analysis
df_no_duplicates.to_excel('CleanedData.xlsx', index=False)
```
