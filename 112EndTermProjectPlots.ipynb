# Plots

```python
import matplotlib.pyplot as plt #importing "matplotlib.pyplot" dictionary with global short name
import seaborn as sns #importing "seaborn" dictionary with global short name
import numpy as np #importing "numpy" dictionary with global short name
import pandas as pd #importing "pandas" dictionary with global short name
```

```python
nba = pd.read_csv('nba_clean.csv') # Reading the dataset
nba.head() # Shows a small piece of the dataset
```

```python
pc = {"PG": "chocolate", "SG": "burlywood", "SF": "lightslategrey", "PF":"steelblue", "C":"darkblue"} # Since the main variable of this project is Pos, 
                                                                                                      # our team decided to use spesific colors for each
                                                                                                      # position in majority of the plots
sns.set(rc={'axes.facecolor':'lavenderblush', 'figure.facecolor':'lightcyan'}) # This line of code arranges background colors of all plots in this project
```

```python
# RQ: Any of the position is prefetable than others?

sns.set_context("notebook", font_scale=1.2) # Changes the default font of the letters on axis
sns.histplot(data=nba, x="Pos", # makes histogram
             stat="percent", # This code makes plot axis count to percentage
             shrink=.8, # makes wider the spaces. Since it is a histogram, at default there is no space between columns
             color="orchid") # color of the columns
plt.title('Percentages of Players Position', # Sets a title
          fontweight="bold", # Makes weight of the title bold
          fontsize=14, # Changes the size of the text
          loc="left") # Locates the title to left
plt.xlabel('') # Sets the name of x axis
plt.show() # Prints the plot
```

```python
# RQ:Which Position Attempts More 3 Point Shot?

sns.set_context("notebook", font_scale=1.2) # Changes the default font of the letters on axis
P3PA=sns.boxplot(data=nba, y="Pos", x="3PA", # Makes boxplot
                 palette=pc, #Arrange the colors to pc ,which we decided in the beginning
                 order=['SG','PG',"SF","PF","C"]) # Orders y axis to interpret easier
P3PA.set(xlabel="Average Attempts Made Per Match", ylabel="") # Sets the names of x and y axis 
plt.title('3 Point Attempts Per Game In Different Positions', # Sets a title
                 fontweight="bold", # Makes weight of the title bold
                 fontsize=14, # Changes the size of the text
                 loc="left") # Sets a title
plt.show() # Prints the plot
```

```python
# RQ:Which Position Makes More Assist?

sns.set_context("notebook", font_scale=1.2) # Changes the default font of the letters on axis
dotc= {"PG": "royalblue", "SG": "darkslategrey", "SF": "w", "PF":"sandybrown", "C":"c"} # A color dictionary to use later
PAST = sns.catplot(data=nba, y="Pos", x="AST", kind="violin", # Makes a catplot (violin plot) 
                   inner=None, # Removes the inner piece of default violin plot
                   palette=pc, #Arrange the colors to pc ,which we decided in the beginning
                   height=5, aspect=1.3, # Arrenges the size of the plot
                   order=['PG','SG',"PF","C","SF"]) # Orders y axis to interpret easier
sns.swarmplot(data=nba, y="Pos", x="AST", # Makes a swarm plot
                   order=['PG','SG',"PF","C","SF"], # Orders y axis to interpret easier
                   palette=dotc) # #Arrange the colors of the dots inside violins to dotc, which we decided earlier in this code group
PAST.set(xlabel="Average Assist Maded Per Game", ylabel="") # Sets the names of x and y axis
plt.title('Number of Total Assists In Different Positions', # Sets a title
                   fontweight="bold", # Makes weight of the title bold
                   fontsize=14,loc="left")  # Changes the size of the text and locates to left
plt.show() # Prints the plot
```

```python
# RQ: Are the average foul scores different between positions?

sns.set_context("notebook", font_scale=1.2) # Changes the default font of the letters on axis
PPF = sns.catplot(data=nba, x="PF", y="Pos", kind="box", # Makes catplot (boxplot)
                  row="Conference", # Draws additional graphs for each of different elements in selected variable
                  palette=pc, #Arrange the colors to pc ,which we decided in the beginning
                  height=2.6, aspect=3) # Arrenges the size of the plot
plt.suptitle('Committed Fouls Per Game In Different Positions', # Sets a subtitle
                  fontweight="bold", # Makes weight of the subtitle bold
                  fontsize=17.5, # Changes the size of text
                  x=0.1,y=1.05,ha="left") # Locates the subtitle
PPF.set_titles(row_template="{row_name}") # Sets titles for each seperate graphs individually with their own name
plt.xlabel('Average Fouls Maded Per Game', # Sets the name of the x axis
                  fontsize=14,  # Changes the size of the text
                  fontweight="bold") # Makes weight of the subtitle bold
PPF.set(ylabel="") # Sets the name of y axis
plt.show() # Prints the plot
```

```python
# RQ:Which Position Grab More Rebounds?

sns.set_context("notebook", font_scale=1.18) # Changes the default font of the letters on axis
PTRB=sns.barplot(data=nba, y="Pos", x="TRB", # Makes a bar plot
                 ci= None, # Removes the coinfidence interval line from the plot
                 order=['C','PF',"SG","SF","PG"], # Orders y axis to interpret easier
                 palette=pc) #Arrange the colors to pc ,which we decided in the beginning
PTRB.set(xlabel="Average Rebounds Per Game", ylabel="") # Sets the names of x and y axis
plt.title('Number of Total Rebounds In Different Positions', # Sets a title
                 fontweight="bold", # Makes weight of the title bold
                 fontsize=14, # Changes the size of the text
                 loc="left") # Locates to left
plt.show() # Prints the plot
```

```python
# RQ:Which Position Makes More Blocks?

sns.set_context("notebook", font_scale=1.18) # Changes the default font of the letters on axis
sns.violinplot(data=nba, x="BLK", y="Pos", # Makes a violin plot
               order=['C','PF',"SG","SF","PG"], # Orders y axis to interpret easier
               palette=pc) #Arrange the colors to pc ,which we decided in the beginning
plt.title('Number of Total Blocks In Different Positions', # Sets a title
               fontweight="bold", # Makes weight of the title bold
               fontsize=14, # Changes the size of the text
               loc="left") # Locates the title to left
plt.xlabel('Average Block Number Per Game ', # Sets the name of x axis
               fontsize=14) # Changes the size of the text
plt.ylabel('') # Sets the name of y axis
plt.show() # Prints the plot
```

```python
# RQ: Does effectiveness ratio of a player increases if that player attempts to do more field goals? Answer: no

sns.set_context("notebook", font_scale=1.2) # Changes the default font of the letters on axis
PFGe = sns.FacetGrid(nba, col="Pos", # An important feature which creates multiple plots
                  height=6, aspect=.45, # Arrenges the size of the plot
                  col_order=['PG','SG',"SF","PF","C"]) # Orders y axis to interpret easier
PFGe.map(sns.scatterplot, "eFG%", "FG", # Makes a scatter plot
                  color="orchid") # Sets a color to the dots
plt.suptitle('Effectiveness Ratio of Field Goals Made', # Adds a subtitle
             fontweight="bold", # Makes weight of the title bold
             fontsize=17.5, # Changes the size of the text
             x=0.071,y=1.024,ha="left") # Locates the subtitle
PFGe.set(xlabel="", ylabel="Average Field Goals Made") # Sets the names of x and y axis
plt.text(x=-1.44, y=-1.96, s='Rates', fontsize=15) # Adds an additional text to the plot
plt.show() # Prints the plot
```
